```python
from astroquery.vsa import Vsa 
import astropy.units as u
import astropy.coordinates as coord

ra = 340.029741
dec = -0.892448
```

```python
result = Vsa.query_region(coord.SkyCoord(ra, dec ,unit=(u.deg, u.deg)),radius=30 * u.arcsec,  programme_id='VHS', database='VHSDR6')
```

```python
df = result.to_pandas()
result
```

```python
images = Vsa.get_image_list(coord.SkyCoord(ra, dec, unit=(u.deg, u.deg)), image_width=180 * u.arcsecond,  programme_id='VHS', database='VHSDR6', frame_type='all', waveband='Ks')
```
